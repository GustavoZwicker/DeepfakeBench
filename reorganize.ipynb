```python
import os
import shutil
from tqdm import tqdm

# ==========================
# PATHS
# ==========================

SOURCE_FRAMES = "C:/Users/Gustavo Zwicker/Documents/DeepfakeBench/dataset_un/test.tar/test/test/imgs"
SOURCE_MASKS = "C:/Users/Gustavo Zwicker/Documents/DeepfakeBench/dataset_un/test.tar/test/test/masks"
LABEL_FILE = "C:/Users/Gustavo Zwicker/Documents/DeepfakeBench/dataset_un/test.tar/test/test/labels.txt"

DEST_ROOT = "datasets/RGB/DDL"

# ==========================
# CREATE STRUCTURE
# ==========================

orig_frames = os.path.join(DEST_ROOT, "ORIGINAL_SEQUENCES", "frames")
orig_masks = os.path.join(DEST_ROOT, "ORIGINAL_SEQUENCES", "masks")

man_frames = os.path.join(DEST_ROOT, "MANIPULATED_SEQUENCES", "frames")
man_masks = os.path.join(DEST_ROOT, "MANIPULATED_SEQUENCES", "masks")

for p in [orig_frames, orig_masks, man_frames, man_masks]:
    os.makedirs(p, exist_ok=True)

# ==========================
# READ LABELS
# ==========================

samples = []

with open(LABEL_FILE) as f:
    for line in f:
        name, label = line.strip().split(",")
        samples.append((name, int(label)))

# ==========================
# COPY FILES
# ==========================

for name, label in tqdm(samples):

    src_frame = os.path.join(SOURCE_FRAMES, name)
    src_mask = os.path.join(SOURCE_MASKS, name)

    if label == 0:
        dst_frame = os.path.join(orig_frames, name)
        dst_mask = os.path.join(orig_masks, name)
    else:
        dst_frame = os.path.join(man_frames, name)
        dst_mask = os.path.join(man_masks, name)

    if os.path.exists(src_frame):
        shutil.copy(src_frame, dst_frame)

    if os.path.exists(src_mask):
        shutil.copy(src_mask, dst_mask)

print("Dataset successfully reorganized.")
```
